**<h1>Heap</h1>**

This is my implementation of the **heap** data structure. The individual functions have been designed to meet the requirements of the task. The methods meet the basic assumptions of the concept of this data structure.

```
class HEAP:
    
    def __init__(self):
        self.data = [] * 20
        self.length = 0
        self.heap_size = 0
    
    def HEAP_INIT(self, tab, tab_length):
        try:
            self.data = tab
            self.length = tab_length
        except:
            print("Too much data!")
            
    def HEAP_MAXIMUM(self):
        return self.data[0]
        
    def MAX_HEAPIFY(self,A, i):
        left = 2 * i + 1
        right = 2 * i + 2
        largest = i
        if left < len(A) and A[left] > A[largest]:
            largest = left
        if right < len(A) and A[right] > A[largest]:
            largest = right
        if largest != i:
            temp = A[largest]
            A[largest] = A[i]
            A[i] = temp
            self.MAX_HEAPIFY(A, largest)

    def BUILD_MAX_HEAP(self):
        for i in range(len(self.data) // 2, -1, -1):
            self.MAX_HEAPIFY(self.data, i)
    
    def MAX_HEAP_INSERT(self, val):
        self.data.append(val)
        self.length += 1
        i = self.length - 1
        for x in range(self.length // 2, -1, -1):
            if self.data[i] > self.data[(i - 1) // 2]:
                temp = self.data[(i - 1) // 2]
                self.data[(i // 2) - 1] = self.data[i]
                self.data[i] = temp
                i = (i // 2) - 1
            else:
                break
            
    def HEAP_EXTRACT_MAX(self):
        result = self.data[0]
        self.data[0] = self.data[self.length - 1]
        del self.data[self.length - 1]
        self.length -= 1
        i = 1
        for x in range(self.length // 2, -1, -1):
            left = (i * 2) - 1
            right = i * 2
            bigger = 0
            lesser = 0
            try:
                if self.data[left] > self.data[right]:
                    bigger = left
                    lesser = right
                if self.data[right] > self.data[left]:
                    bigger = right
                    lesser = left
            except IndexError:
                break
            if self.data[i - 1] < self.data[bigger]:
                temp = self.data[bigger]
                self.data[bigger] = self.data[i - 1]
                self.data[i - 1] = temp
                i = bigger + 1
                continue
            elif self.data[i - 1] < self.data[lesser]:
                temp = self.data[lesser]
                self.data[lesser] = self.data[i - 1]
                self.data[i - 1] = temp
                i = lesser + 1
        return result
        
    def INCREASE_KEY(self,index, new_value):
        idx = index - 1
        self.data[idx] = new_value
        for x in range(self.length // 2, -1, -1):
            if self.data[idx] > self.data[idx // 2]:
                temp = self.data[idx // 2]
                self.data[idx // 2] = self.data[idx]
                self.data[idx] = temp
                idx = idx // 2
            else:
                break
        
    def WRITE(self, *args):
        if args == ():
            result = ""
            for x in range(0,len(self.data)):
                result = result + str(self.data[x]) + " "
            print(result)
        else:
            print(str(args)[1:-2].strip("'").lower())
        

tab =[4,1,3,2,16,9,10,14,8,7]
instance = HEAP()
instance.HEAP_INIT(tab,len(tab))
instance.BUILD_MAX_HEAP()
instance.WRITE()
instance.MAX_HEAP_INSERT(15)
instance.WRITE()
instance.WRITE(instance.HEAP_MAXIMUM())
instance.WRITE()
instance.WRITE(instance.HEAP_EXTRACT_MAX())
instance.WRITE()
instance.INCREASE_KEY(4,19)
instance.WRITE()
```
