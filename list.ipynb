**<h1>List</h1>**

This is my implementation of the **list** data structure. The individual functions have been designed to meet the requirements of the task. The methods meet the basic assumptions of the concept of this data structure.

```

class Node:
    def __init__(self):
        self.key = None
        self.next = None
        self.prev = None
        
class List:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def INSERT(self, var):
        n = Node()
        n.key = var
        if self.head == None:
            self.head = n
            self.tail = n
        else:
            n.next = self.head
            self.head.prev = n
            self.head = n
            
    def INSERT_AFTER(self,var,aft):
        n = Node()
        n.key = var
        x = self.head
        
        while x != None:
            if x.key == aft:
                break
            x = x.next
        
        temp = x.next
        x.next = n
        temp.prev = n
        n.prev = x
        n.next = temp
        
    def INSERT_BEFORE(self,var,bef):
        n = Node()
        n.key = var
        x = self.head
        
        while x != None:
            if x.key == bef:
                break
            x = x.next
        
        if x.prev != None:
            temp = x.prev
            x.prev = n
            temp.next = n
            
            temp = None
        else:
            self.INSERT(var)
        
            
    def WRITE(self, *args):
        if args == ():
            x = self.head
            result = ""
            while x != None:
                result = result + str(x.key) + " "
                x = x.next
            print(result)
        else:
            print(str(args)[1:-2].strip("'").lower())
        
    def SEARCH(self, var):
        x = self.head
        
        while x != None:
            if x.key == var:
                return x.key
            x = x.next
        return -1
    
    def CLEAR(self):
        while self.head != None:
            temp = self.head
            self.head = self.head.next
            del temp
        
    def DELETE(self, var):
        
        current = self.head 
        prev = None
        
        while current and current.key != var:
            prev = current
            current = current.next
        if prev is None:
            self.head = current.next
        elif current:
            prev.next = current.next
            current.next = None
            
        current = None
        prev = None
            

instance = List()
instance.WRITE()
instance.INSERT(1)
instance.INSERT(2)
instance.WRITE()
instance.INSERT(3)
instance.WRITE(instance.SEARCH(2))
instance.WRITE(instance.SEARCH(4))
instance.INSERT_AFTER(4,2)
instance.WRITE()
instance.DELETE(2)
instance.WRITE()
instance.DELETE(5)
instance.WRITE()
instance.INSERT_BEFORE(5,3)
instance.WRITE()
instance.CLEAR()
instance.WRITE()
```
